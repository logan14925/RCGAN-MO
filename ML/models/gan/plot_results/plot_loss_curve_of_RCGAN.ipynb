```python
import pandas as pd
import matplotlib.pyplot as plt

# 从 CSV 文件加载数据
csv_file_path = 'E:/01_Graduate_projects/Cellular_structures/code_and_pth/Cellular_Project/ML/models/gan/RCGAN_train_results.csv'
df = pd.read_csv(csv_file_path)

# 提取需要的列
epochs = df['Epoch']
D_Losses = df['D_Loss']
G_Losses = df['G_Loss']
R_Losses = df['R_Loss']
D_real_accuracies = df['D_Real_Accuracy']
D_fake_accuracies = df['D_Fake_Accuracy']

# 设置全局字体和大小
plt.rcParams.update({
    'font.weight': 'bold',            # 加粗字体
    'axes.labelweight': 'bold',       # 加粗坐标轴标签
    'axes.titlesize': 26,            # 增加标题的字体大小
    'axes.labelsize': 18,            # 增加坐标轴标签字体大小
    'xtick.labelsize': 12,           # 增加X轴刻度字体大小
    'ytick.labelsize': 12,           # 增加Y轴刻度字体大小
    'axes.titleweight': 'bold'       # 标题加粗
})

# 绘制 D Loss 曲线
plt.figure(figsize=(8, 6))
plt.plot(epochs, D_Losses, label="D Loss", color='red')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Discriminator Loss", pad=10)
plt.tight_layout()
plt.show()

# 绘制 G Loss 曲线
plt.figure(figsize=(8, 6))
plt.plot(epochs, G_Losses, label="G Loss", color='blue')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Generator Loss", pad=10)
plt.tight_layout()
plt.show()

# 绘制 R Loss 曲线
plt.figure(figsize=(8, 6))
plt.plot(epochs, R_Losses, label="R Loss", color='green')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Regressor Loss", pad=10)
plt.tight_layout()
plt.show()

# 绘制 Discriminator Accuracy 曲线
plt.figure(figsize=(8, 6))
plt.plot(epochs, D_real_accuracies, label="Real Accuracy", color='blue')
plt.plot(epochs, D_fake_accuracies, label="Fake Accuracy", color='red')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Discriminator Accuracy", pad=10)
plt.legend()
plt.tight_layout()
plt.show()

print("Final models have been saved.")
```
